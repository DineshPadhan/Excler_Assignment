Exercise 1: Prime Numbers

Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python
num = int(input("Enter the Number: "))
if num > 1:
    for i in range(2, int(num/2)+1):
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
```

Exercise 2: Product of Random Numbers

Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.

```python
import random
num1 = random.randint(1, 20)
num2 = random.randint(1, 20)

user_product = int(input(f"What is the product of {num1} and {num2}? "))

print(f"You entered: {user_product}")

if user_product == num1 * num2:
    print("Correct! Well done.")
else:
    print(f"Incorrect. The correct answer is {num1 * num2}.")
```

Exercise 3: Squares of Even/Odd Numbers

Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200. Choose either even or odd numbers and document your choice in the code.

```python
print("Choose odd or even to print the squares of all even/odd numbers within the range of 100 to 200")

choice = input("Enter your choice (odd/even): ").strip().lower()

if choice == "even":
    numbers = range(100, 201, 2)
else:
    numbers = range(101, 200, 2)

squares = [n**2 for n in numbers]
print("The squares of", choice, "numbers between 100 and 200 are:", squares)
```

✅ Problem Statement (Question): Hospital Billing System
A hospital charges its patients based on the following:
•	Room charges per day: ₹2,000
•	Doctor consultation fee (flat): ₹1,500
•	Lab test charges: Based on the number of tests taken. Each test costs ₹300
•	Medicine charges: Total cost of medicines provided
•	Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
Write a Python program to calculate the final bill for a patient who:
•	Stayed for 4 days
•	Had 3 lab tests
•	Had medicine charges of ₹2,400

```python
import time

def calculate_bill(Room_Charge, Doctor_Fee, lab_charge, days, lab_tests, medicines):
    print("Calculating total bill...")
    time.sleep(2)

    total_room_charge = Room_Charge * days
    print("Total room charge:", total_room_charge)

    total_lab_charge = lab_charge * lab_tests
    print("Total lab charge:", total_lab_charge)

    total_bill = total_room_charge + Doctor_Fee + total_lab_charge + medicines
    print("Total bill before discount (if applicable):", total_bill)

    if total_bill > 10000:
        print("You are eligible for a discount of 10% on your total bill.")
        total_bill = discount(total_bill)
        print(f"The total bill after discount: {total_bill}")

def discount(total_bill):
    if total_bill > 10000:
        discount = total_bill * 0.10
        total_bill -= discount
        print(f"A discount of {discount} has been applied.")
    return total_bill

def main():
    Room_Charge = 2000
    Doctor_Fee = 1500
    lab_charge = 300
    days = int(input("Enter the number of days you stayed in the hospital: "))
    print("Total days stayed:", days)

    lab_tests = int(input("Enter the number of lab tests you had: "))
    print("Total lab tests:", lab_tests)

    medicines = float(input("Enter the total cost of medicines: "))
    print("Total cost of medicines:", medicines)
    
    calculate_bill(Room_Charge, Doctor_Fee, lab_charge, days, lab_tests, medicines)

main()
```
