#### 1. State the Hypotheses statement:

It is a statistical data that is used in making statistical decision using experimental data.
It is basically an assumption that we make about the population parameter.
It is a 2 step process:
1. H0: Null Hypotheses, where data is following normal distribution
2. H1: Alternative Hypotheses, where data is not following normal distribution.

```python
# 2. Calculate the Test Statistic:
# operating cost per week calculation formula is = $1000 + $5X where X repersent the number of units produced in a week
x_bar = 3050.0      
mu_theoretical = 1000.0 + 5.0 * 600.0
sigma = 5.0 * 25.0
n = 25
alpha = 0.05
```

```python
# Calculate the Z test statistic
import math
from scipy.stats import norm

# Z test statistic
z = (x_bar - mu_theoretical) / sigma / math.sqrt(n)
print(f"Z test statistic: {z:.4f}")

critical_value = norm.ppf(1 - alpha)
print(f"Critical value (one-tailed): {critical_value:.4f}")
```

```python
# 3. Determine the Probability (P-value)\
p_value = 1 - norm.cdf(z)
print(f"P-value: {p_value:.10f}")

# Since the z-statistic is negative and we're testing if costs are HIGHER,
# this means the sample mean is actually LOWER than expected
print(f"The sample mean (${x_bar}) is actually lower than the theoretical mean (${mu_theoretical})")
```

```python
# 4. Make a Decision:
# Decision rule for one-tailed test (right-tail):
# Reject H0 if Z > critical_value OR if p-value < α
if z > critical_value or p_value < alpha:
    print("There is sufficient evidence to support the claim that weekly operating costs are higher than the model suggests.")
else:
    print("There is insufficient evidence to support the claim that weekly operating costs are higher than the model suggests.")
```

```python
# 5. Conclusion:
# There is NO strong statistical evidence (at alpha = 0.05 level) to support the restaurant owners' claim that weekly operating costs are higher than what the cost model W = $1,000 + $5X predicts.
#  The sample data shows that the actual mean cost is significantly LOWER than the theoretical mean cost. 
# This suggests the cost model may actually be OVERESTIMATING costs, not underestimating them as the restaurant owners claimed.

# The extremely small p-value indicates this difference is highly significant.
# The large negative Z-score shows the sample mean is much lower than expected.
# This contradicts the restaurant owners' claim of higher costs.
```
