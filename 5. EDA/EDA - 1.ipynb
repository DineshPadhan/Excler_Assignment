```python
# 1.	Data Cleaning and Preparation:
#     * Load the dataset into a data frame or equivalent data structure.
#     * Handle missing values appropriately (e.g., imputation, deletion).
#     * Identify and correct any inconsistencies in data types (e.g., numerical values stored as strings).
#     * Detect and treat outliers if necessary.

import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('./Cardiotocographic.csv')
df
```

```python
df.info()
```

```python
# Handle missing values appropriately (e.g., imputation, deletion).
print("Before handling missing values:\n",df.isnull().sum())
dfcols = df.columns
for col in dfcols:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("After handling missing values:\n",df.isnull().sum())
```

```python
# Identify and correct any inconsistencies in data types (e.g., numerical values stored as strings).
print("Before:",df.dtypes)
# Convert columns to appropriate data types if necessary
for col in df.columns:
    df[col] = df[col].apply(pd.to_numeric, errors='coerce')
print("After:",df.dtypes)
```

```python
# Detect outliers
import matplotlib.pyplot as plt
dfcols=df.columns

dashboard,ax=plt.subplots(5,3,figsize=(30,15))
for i in range(5):
    for j in range(3):
        if i==4 and j==2:
            break
        df[dfcols[0]].plot(kind='box',ax=ax[i,j], vert=False)
        ax[i,j].set_title(dfcols[0], fontsize=20)
        plt.subplots_adjust(hspace=0.5)
        dfcols=dfcols[1:]
plt.show()
```

```python
# Treat outliers if necessary.
# for i in range(2):
for col in df.columns:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    LW = Q1 - 1.5 * IQR
    UW = Q3 + 1.5 * IQR
    df[df[col] > UW] = UW
    df[df[col] < LW] = LW
```

```python
# Detect outliers again to confirm
dfcols=df.columns

dashboard,ax=plt.subplots(5,3,figsize=(30,15))
for i in range(5):
    for j in range(3):
        if i==4 and j==2:
            ax[i,j].axis('off')
            break
        df[dfcols[0]].plot(kind='box',ax=ax[i,j], vert=False)
        ax[i,j].set_title(dfcols[0], fontsize=20)
        plt.subplots_adjust(hspace=0.5)
        dfcols=dfcols[1:]
plt.show()
```

```python
# 2.	Statistical Summary:
#     * Provide a statistical summary for each variable in the dataset, including measures of central tendency (mean, median) and dispersion (standard deviation, interquartile range).
#     * Highlight any interesting findings from this summary.
print(df.describe())        #it provides count, mean, std, min, 25%, 50%, 75%, max
print("Median:\n",df.median())         #it provides median of each column
```

```python
# 3.	Data Visualization:
#     * Create histograms or boxplots to visualize the distributions of various numerical variables.
#     * Use bar charts or pie charts to display the frequency of categories for categorical variables.
#     * Generate scatter plots or correlation heatmaps to explore relationships between pairs of variables.
#     * Employ advanced visualization techniques like pair plots, or violin plots for deeper insights.

# * Create histograms or boxplots to visualize the distributions of various numerical variables.
import seaborn as sns

dashboard,ax=plt.subplots(5,3,figsize=(30,15))
for i in range(5):
    for j in range(3):
        idx = i * 3 + j
        if idx >= len(df.columns):
            ax[i, j].axis('off')
            continue
        col = df.columns[idx]
        sns.histplot(df[col], ax=ax[i, j], bins=30, kde=True, edgecolor='black')
        ax[i, j].set_title(f'Histogram of {col}', fontsize=20)
        ax[i, j].set_xlabel(col, fontsize=15)
        ax[i, j].set_ylabel('Frequency', fontsize=15)
plt.subplots_adjust(hspace=1)

# Boxplots
dashboard,ax=plt.subplots(5,3,figsize=(30,15))
dfcols=df.columns
for i in range(5):
    for j in range(3):
        if i == 4 and j == 2:
            ax[i, j].axis('off')
            continue
        col = dfcols[0]
        sns.boxplot(x=df[col], ax=ax[i, j])
        ax[i, j].set_title(f'Boxplot of {col}', fontsize=20)
        ax[i, j].set_xlabel(col, fontsize=15)
        ax[i, j].set_ylabel('Value', fontsize=15)
        dfcols = dfcols[1:]
plt.subplots_adjust(hspace=1)
```

```python
# * Generate scatter plots or correlation heatmaps to explore relationships between pairs of variables.

# Correlation Heatmap
plt.figure(figsize=(15, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation Heatmap', fontsize=20)
plt.show()

# Scatter Plots 
dashboard, ax = plt.subplots(5, 3, figsize=(30, 15))
num_plots = 0
max_plots = 15  # 5x3 grid

for i in range(len(df.columns)-1):
    for j in range(i+1, len(df.columns)):
        if num_plots >= max_plots:
            break
        row = num_plots // 3
        col = num_plots % 3
        df.plot(kind='scatter', x=df.columns[i], y=df.columns[j], ax=ax[row, col], title=f'Scatter Plot of {df.columns[i]} vs {df.columns[j]}')
        num_plots += 1

plt.tight_layout()
plt.show()
```
